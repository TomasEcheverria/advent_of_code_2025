```python
voltagee = []
for line in parsed_real_data:
    max=""
    for index in range(len(line)):
        first = line[index]
        for second_index in range(index + 1, len(line)):
            second = line[second_index]
            candidate = first + second
            if candidate > max:
                print("New max found:", candidate)
                max = candidate
    voltagee.append(max)
print(voltagee)
sum([int(c) for c in voltagee])
```
