```python
arr = [1024,512,256,128,64,32,16,8,4,2,1]
from collections import defaultdict
occurences = defaultdict(list)
from collections import Counter
for elem in arr:
    count = Counter(bin(elem))['1']
    occurences[count].append(elem)

    

solution = []
for count in sorted(occurences.keys()):
    solution.extend(sorted(occurences[count]))
print(solution)
```
