```python
def id_has_invalid_format(id):
    if id.startswith("0"):
        print(f"{id} is invalid cause it starts with 0")
        return True
        
    if len(id) % 2 != 0 : 
        return False
    return False

def filter_id_between_range(id_range):
    invalid_id_list = []
    fist = int(id_range[0])
    last = int(id_range[1])
    for id in range(fist, last +1):
        id = str(id)
        if id_has_invalid_format(id):
            invalid_id_list.append(id)
            continue
        amount_of_repetitions = len(id) // 2
        for i in range(1,  amount_of_repetitions + 1):
            number = id[i:] * 2 
            if number == id:
                #print(f"{number} twice repeated forms invalid : {id}")
                invalid_id_list.append(id)
    print(f"invalid ids are: {invalid_id_list} in range {id_range}")
    return invalid_id_list

invalid_ids = []
[invalid_ids.extend(filter_id_between_range(id_range= data[i])) for i in range(iterations)]
sum([int (e) for e in invalid_ids])


```
