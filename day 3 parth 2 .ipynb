```python
sample = ["987654321111111", "811111111111119", "234234234234278", "818181911112111"]
import requests
url = "https://adventofcode.com/2025/day/3/input"
session = "53616c7465645f5f4fc9ce9761c1c5fefa47aab8a7c9c4dd1be33644e6afa1c733f4ce8b2890fb1b385cb39624b88c1dc19d9d0cc8e7d640b18dfed7da1be127"
cookies = {"session": session}
response = requests.get(url, cookies=cookies)
parsed_real_data = (response.text).splitlines()
k = 12 
```

```python
def max_joltage_12(bank: str, k: int = 12) -> str:
    stack = []
    to_remove = len(bank) - k

    for digit in bank:
        while stack and to_remove > 0 and stack[-1] < digit:
            stack.pop()
            to_remove -= 1
        stack.append(digit)

    return ''.join(stack[:k])
```

```python
sum([int (c) for c in [max_joltage_12(bank, k) for bank in parsed_real_data]])
```
